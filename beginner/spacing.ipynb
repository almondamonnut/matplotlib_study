# 1. Why need spacing?

อย่างแรก มาดูกันว่าทำไมเราถึงอยากเซ็ตค่าความห่าง

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt

# create data
x = np.array([1, 2, 3, 4, 5])

# making subplots object
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)
plt.show()
```

แล้วแบบรูปภาพล่ะ?

```python
from PIL import Image
img1 = np.asarray(Image.open('1.jpg'))
img2 = np.asarray(Image.open('2.jpg'))
img3 = np.asarray(Image.open('3.jpg'))
img4 = np.asarray(Image.open('4.jpg'))

# making subplots object
fig, ax = plt.subplots(2, 2)

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 1].imshow(img2)
ax[1, 0].imshow(img3)
ax[1, 1].imshow(img4)
```

จะเห็นว่าแบบกราฟนั้น space น้อยไป 
ส่วนแบบภาพนั้น space มากไป

# 2. Set spacing between subplots

### กราฟ

#### แบบไม่จัดรูปแบบ

```python
### แบบไม่มี tight_layout()

# create data
x = np.array([1, 2, 3, 4, 5])

# making subplots object
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)
plt.show()
```

#### แบบจัดรูปแบบ

ใช้ parameter wspace เพื่อปรับ width space คือ ความห่างระหว่าง subplot ซ้ายขวา 


และใช้ parameter hspace = 0.4 เพื่อปรับ height space คือ ความห่างระหว่าง subplot บนล่าง 

```python

# making subplots object
fig, ax = plt.subplots(2, 2)

# เซ็ตค่าความห่าง
plt.subplots_adjust(wspace = 0.4, hspace = 0.4) 

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)
plt.show()
```

### ภาพ

#### แบบไม่จัดรูปแบบ

```python

# making subplots object
fig, ax = plt.subplots(2, 2)

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 1].imshow(img2)
ax[1, 0].imshow(img3)
ax[1, 1].imshow(img4)
plt.show()
```

#### แบบจัดรูปแบบ

```python

# making subplots object
fig, ax = plt.subplots(2, 2)

# เซ็ตค่าความห่าง
plt.subplots_adjust(wspace = -0.4, hspace = 0) 

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 0].axis('off')
ax[0, 1].imshow(img2)
ax[0, 1].axis('off')

ax[1, 0].imshow(img3)
ax[1, 0].axis('off')
ax[1, 1].imshow(img4)
ax[1, 1].axis('off')

plt.show()
```
