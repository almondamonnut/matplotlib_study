# 1. Why need spacing?

อย่างแรก มาดูกันว่าทำไมเราถึงอยากเซ็ตค่าความห่าง

```python
# import packages
import numpy as np
import matplotlib.pyplot as plt

# create data
x = np.array([1, 2, 3, 4, 5])

# making subplots object
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)
plt.show()
```

แล้วแบบรูปภาพล่ะ?

```python
from PIL import Image
img1 = np.asarray(Image.open('1.jpg'))
img2 = np.asarray(Image.open('2.jpg'))
img3 = np.asarray(Image.open('3.jpg'))
img4 = np.asarray(Image.open('4.jpg'))

# making subplots object
fig, ax = plt.subplots(2, 2)

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 1].imshow(img2)
ax[1, 0].imshow(img3)
ax[1, 1].imshow(img4)
```

จะเห็นว่าแบบกราฟนั้น space น้อยไป 
ส่วนแบบภาพนั้น space มากไป

# 2. Set spacing between subplots

### 2.1 ใช้ tight_layout() ในการเซ็ตความห่าง

tight_layout() method จะเซ็ตระยะห่างระหว่าง subplots ที่เหมาะสมโดยอัตโนมัติ

```python

### แบบไม่มี tight_layout()

# create data
x = np.array([1, 2, 3, 4, 5])

# making subplots object
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)
plt.show()
```

```python

### แบบมี tight_layout()

# create data
x = np.array([1, 2, 3, 4, 5])

# making subplots object
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x**2)
ax[1, 1].plot(x, x**3)

fig.tight_layout()
plt.show()
```

สำหรับรูปภาพบ้างล่ะ?

```python

### แบบไม่มี tight_layout()

# making subplots object
fig, ax = plt.subplots(2, 2)

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 1].imshow(img2)
ax[1, 0].imshow(img3)
ax[1, 1].imshow(img4)
plt.show()
```

```python

### แบบไม่มี tight_layout()

# making subplots object
fig, ax = plt.subplots(2, 2)
fig.tight_layout()

# use each axes object to show each image
ax[0, 0].imshow(img1)
ax[0, 1].imshow(img2)
ax[1, 0].imshow(img3)
ax[1, 1].imshow(img4)
plt.show()
```

จะเห็นว่าไม่เวิร์คจ้า แบบภาพยังไงก็ spacing เยอะไปอยู่ดี
